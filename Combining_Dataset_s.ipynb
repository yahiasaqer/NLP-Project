```python
import pandas as pd

# List of file paths for the six Excel files
file_paths = [
    '/content/drive/MyDrive/split_file_1.xlsx',
    '/content/drive/MyDrive/split_file_2.xlsx',
    '/content/drive/MyDrive/split_file_3.xlsx',
  #  '/content/drive/MyDrive/augmented_stress_file.xlsx',
  #  '/content/drive/MyDrive/augmented_depression_file.xlsx',
  #  '/content/drive/MyDrive/augmented_suicidal_file.xlsx',
    # '/content/drive/MyDrive/augmented_anxiety_file.xlsx'
]

# Read each Excel file into a DataFrame
dataframes = [pd.read_excel(file) for file in file_paths]

# Concatenate all DataFrames vertically
combined_df = pd.concat(dataframes, ignore_index=True)

# Save the combined DataFrame to a new Excel file
combined_df.to_excel('original_dataset.xlsx', index=False)

print("Excel files have been combined and saved.")
```

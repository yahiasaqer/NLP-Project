```python
exceptional_words = [
    # Names starting with 'س'
    "سنبل", "سنان", "سنام", "سند", "سنور", "سنوار", "ستر", "ستار", "ستين",
    "سنح", "سن", "سيف", "سند", "سيطر", "سأم", "سأل", "ستون", "سنوريات",
    "سيد", "ستير", "سنابل", "سياق", "سير", "سيما", "سأف", "سأمان",
    "سياسي", "سيجار", "سيبويه", "سني", "سنجاب", "سنبوق", "سنابك",
    "ستينيات", "سجائر", "ستائر", "سيار", "سيرك", "سيلان", "سيل", "سنجاب", "سنوات",
"سألتني",
    "سألتنا",
    "سألتكم",
    "سألتكما",
    "سألتكن",
    "سأله",
    "سألهما",
    "سألههم",
    "سألهن",
    "سيئ",
    "وسيئ",
"سألت",
"سألنا",
"سيء",
"سيئان",
"سئين",
"سيئات",
"سندان",
"ستار",
"ستائر",
"سيام",
"سيدان",
"سيرا",
"سيروا",
"سيرن",
"سنام",
"سنامان",
"سنامين",
"وسأل",
"وسند",
"وسار",
"وساء",
"وسيل",
"وسأم",
"وسيط",
"وسيع",
"وسيق",
"وسيلة",
"وسيلتان",
"وسلتين",
"وسيلات",
"وسائل",
"وسيطان",
"وسيطين",
"وسيطون",
"وسيم",
"وسيمة",
"وسيمين",
"وسيمان",
"وسيمون",
"وسيمات",
"وسيطات"
"وسيطة",
"وستر",
"وسنت",
"وسنة",
"وسنن",
"وسني",
"وسنيين",
"وسنيات",
"وسنيون",
"سنيين",
"سنيات",
"سنيون",
"سيناريو",
 "وسنبل", "وسنان", "وسنام", "وسند", "وسنور", "وسنوار", "وستر", "وستار", "وستين",
    "وسنح", "وسن", "وسيف", "وسند", "وسيطر", "وسأم", "وسأل", "وستون", "وسنوريات",
    "وسيد", "وستير", "وسنابل", "وسياق", "وسير", "وسيما", "وسأف", "وسأمان",
    "وسياسي", "وسيجار", "وسيبويه", "وسني", "وسنجاب", "وسنبوق", "وسنابك",
    "وستينيات", "وسجائر", "وستائر", "وسيار", "وسيرك", "وسيلان", "وسيل", "وسنجاب", "وسنوات",
"وسألتني",
"وسناريو",
    "سألتنا",
    "وسألتكم",
    "وسألتكما",
    "وسألتكن",
    "وسأله",
    "وسألهما",
    "وسألههم",
    "وسألهن",
"وسألت",
"وسألنا",
"وسيء",
"وسيئان",
"وسئين",
"وسيئات",
"وسندان",
"وستار",
"وستائر",
"وسيام",
"وسيدان",
"وسيرا",
"وسيروا",
"وسيرن",
"وسنام",
"وسنامان",
"وسنامين",
"سنبلة",
"سنبلتين",
"سنبلتان",
"سندين",
"ستارين",
"ستاران",
"سيطرا",
"سيطرتا",
"سيطروا",
"سيطرن",
"سيطرت",
"سيطرنا",
"سيفان",
"سيفين",
"وسنة",
"سنة",
"سنجابين",
"سنجابان",
"سناجب",
"سيدان",
"سيدين",
"سيدتين",
"سيدتان",
"سيدات",
"سياسة",
"سياسيين",
"سياسيان",
"سياسية",
"سيدة",
"سياسيات",
"سياسيتان",
"سياسيتين",
"سياسيون",
 "وسنبلة",
    "وسنبلتين",
    "وسنبلتان",
    "وسندين",
    "وستارين",
    "وستاران",
    "وسيطرا",
    "وسيطرتا",
    "وسيطروا",
    "وسيطرن",
    "وسيطرت",
    "وسيطرنا",
    "وسيفان",
    "وسيفين",

    "وسنة",
    "وسنجابين",
    "وسنجابان",
    "وسناجب",
    "وسيدان",
    "وسيدين",
    "وسيدتين",
    "وسيدتان",
    "وسيدات",
    "وسياسة",
    "وسياسيين",
    "وسياسيان",
    "وسياسية",
    "وسيدة",
    "وسياسيات",
    "وسياسيتان",
    "وسياسيتين",
"سيء",
"وسيء",
"سيئا",
"وسيئا",
]


attached_pronouns = ["ه", "ها","هم","هن","ك","كي","كن","كم","كما", "ت","نا","تما","تم","تن","تي", "ات","ان", "ين", "ون"]
# Create the concatenated strings
concatenated_strings = [word + suffix for word in exceptional_words for suffix in attached_pronouns]

# Merge both lists
prohibited_words = exceptional_words + concatenated_strings


def replace_sin_with_rah(sentence):
    """
    Replace 'سين التسويف' with 'راح' if found at the beginning of present verbs in a sentence.

    Present verbs are known to start with أ, ي, ت, or ن.
    """

    # Define prefixes and suffixes to identify nouns
    noun_suffixes = ('ة', "ً", "ٍ")  # Add more suffixes if needed
    noun_prefixes = ('ال',)  # Add more prefixes if needed
    prepositions = ('من', 'إلى', 'على', 'في', 'عن', 'مع', 'ب', 'ك', "الى")  # Common prepositions
    interjections = ('أيا', 'يا')  # Common interjections

    # Define a regular expression pattern to match present verbs starting with أ, ي, ت, ن
    pattern = re.compile(r'^(و?س)(أ|ي|ت|ن)')

    # Split the sentence into words
    words = sentence.split()

    # Process each word in the sentence
    processed_words = []
    for word in words:
        # Check if the word is preceded by any of the specified prepositions or interjections
        if any(word.startswith(prep) or word.startswith(interj) for prep in prepositions for interj in interjections):
            processed_words.append(word)
            continue

        # Check if the word starts with 'ا' or 'أ' and is more than 3 letters long or if it starts with present verb patterns
        if (word.startswith(('أ', 'ي', 'ن', 'ت')) and len(word) > 3) or pattern.match(word):
            # Check if the word is not a noun based on its suffixes or prefixes and not an Arabic name
            if not word.endswith(noun_suffixes) and not word.startswith(noun_prefixes) and word not in prohibited_words:
                # Replace 'سين التسويف' with 'راح' at the beginning of the verb
                processed_word = re.sub(r'^(و?)س(أ|ي|ت|ن)', r'\1راح \2', word)
                processed_words.append(processed_word)
            else:
                processed_words.append(word)
        else:
            processed_words.append(word)

    # Reassemble the sentence
    return ' '.join(processed_words)

# Example usage
sentence = "سأذهب إلى السوق وسأشتري بعض الأشياء سأل وسأل"
result = replace_sin_with_rah(sentence)
print(result)  # Output: راح أذهب إلى السوق وراح أشتري بعض الأشياء
```

```python

def process_b_after_rah(text):
    """
    Processes words preceded by 'راح' according to the following rules:
    1. Removes the letter 'ب' at the beginning of the word if preceded by 'راح'.
    2. If 'ب' is not followed by 'ت', 'ن', or 'ي', replace it with 'أ', but only if preceded by 'راح'.
    """
    def replace_match(match):
        word = match.group(2)
        if len(word) > 1 and word[0] == 'ب':
            if word[1] in 'تني':  # Check the letter following 'ب'
                return f"راح {word[1:]}"  # Remove 'ب'
            else:
                return f"راح أ{word[1:]}"  # Replace 'ب' with 'أ'
        return match.group(0)

    # Match 'راح' followed by a space and a word starting with 'ب'
    pattern = r'(راح\s)(ب\w+)'
    result = re.sub(pattern, replace_match, text)
    return result

# Example usage
text = "راح بيلعب وراح بنجح وراح بيقدر وراح بقرا"
processed_text = process_b_after_rah(text)
print(processed_text)
```

```python
# Initialize the dialect replacements dictionary
oreed_replacements = {}

# Define the dialect replacements with connected pronouns
keys_for_bidi = ["أريد", "اريد",  "أبغى", "ابغى"]
for key in keys_for_bidi:
    oreed_replacements[key] = "بدي"

keys_for_bidna = ["نريد", "نبغى"]
for key in keys_for_bidna:
    oreed_replacements[key] = "بدنا"

keys_for_bidha = ["تريد", "تبغى", ]
for key in keys_for_bidha:
    oreed_replacements[key] = "بدها"

keys_for_bidhom = ["يريدون", "يبغون", "يريدان", "يبغان"]
for key in keys_for_bidhom:
    oreed_replacements[key] = "بدهم"

keys_for_bidho = ["يريد", "يبغى", "يبغى"]
for key in keys_for_bidho:
    oreed_replacements[key] =  "بدو"

keys_for_bidhon = ["يردن",  "يبغين"]
for key in keys_for_bidhon:
    oreed_replacements[key] = "بدهن"

keys_for_bidak = ["تريد", "تبغى"]
for key in keys_for_bidak:
    oreed_replacements[key] = "بدك"
keys_for_bidkom = ["تريدون", "تبغون", ]
for key in keys_for_bidkom:
    oreed_replacements[key] = "بدكم"

keys_for_bidkon = ["تردن",  "تبغن"]
for key in keys_for_bidkon:
    oreed_replacements[key] = "بدكم"
keys_for_kanbid = ["أراد"]
for key in keys_for_kanbid:
    oreed_replacements[key] = "كان بدو"
keys_for_kanbidha = ["أرادت"]
for key in keys_for_kanbidha:
    oreed_replacements[key] = "كان بدها"

keys_for_kanbidi = ["أردت", "بغيت"]
for key in keys_for_kanbidi:
    oreed_replacements[key] = "كان بدي"
# Define the pronouns and their replacements
pronoun_replacements = {
    "ه": " إياه",
    "ها": " إياها",
    "هما": " إياهم",
    "هم": " إياهم",
    "هن": " إياهن",
    "ك": " إياك",
    "كي": " إياك",
    "كما": " إياكم",
    "كم": " إياكم",
    "كن": " إياكم",
    "ني":"إياني",
    "نا":" إيانا",


}

def replace_pronouns_in_verb(phrase, verb_key):
    """
    Replace pronouns only in the segment where a specific verb is found.
    """
    # Replace connected pronouns only in the context of the replaced verb
    for pronoun, replacement in pronoun_replacements.items():
        phrase = phrase.replace(verb_key + pronoun, verb_key + replacement)
    return phrase

def translate_wanting_verbs(phrase):
    """
    Translate standard Arabic verbs to Palestinian dialect considering connected pronouns
    only for verbs found in oreed_replacements.
    """


    final_phrase = phrase
    for key, replacement in oreed_replacements.items():
        if key in final_phrase :
            # Replace the verb with its Palestinian dialect equivalent
            final_phrase = final_phrase.replace(key, replacement)
            # Replace pronouns only for the verb that has been replaced
            final_phrase = replace_pronouns_in_verb(final_phrase, replacement)

    # Clean up any stray " إيا "
    final_phrase = final_phrase.replace(" إيا ", "")

    return final_phrase
# Example usage
print(translate_wanting_verbs('أريدكن'))  # Output: بدي إياك
print(translate_wanting_verbs('نريدها'))  # Output: بدنا إياها
print(translate_wanting_verbs('يريدونك'))  # Output: بدهم إياك
print(translate_wanting_verbs('تريدهم'))  # Output: بدها إياهم
print(translate_wanting_verbs('أرادها'))
print(translate_wanting_verbs('أردته'))
```

```python


def remove_arabic_punctuation(text):
    # Arabic punctuation marks and additional symbols to remove
    arabic_punctuation = (
        "\u060C\u061B\u061F\u066A\u066B\u066C\u066D\u06D4"
        "\u06D6\u06D7\u06D8\u06D9\u06DA\u06DB\u06DC\u06DD"
        "\u06DE\u06DF\u06E0\u06E1\u06E2\u06E3\u06E4\u06E5"
        "\u06E6\u06E7\u06E8\u06E9\u06EA\u06EB\u06EC\u06ED"
        "\u0021\u002E\u002C\u003F\u003B\u003A\u0028\u0029"
        "\u005B\u005D\u007B\u007D\u003C\u003E\u0022\u0027"
        "\u2018\u2019\u201C\u201D\u00AB\u00BB\u2039\u203A"
    )

    # Escape characters that need escaping in regex
    escaped_punctuation = re.escape(arabic_punctuation)

    # Create a regex pattern for punctuation
    pattern = f"[{escaped_punctuation}]"

    # Remove punctuation from the text
    cleaned_text = re.sub(pattern, ' ', text)

    return cleaned_text

# Example usage
text = "مرحباً، كيف حالك؟ أريد!. أن أزورك! (متى؟) [غداً] {أكيد}"
cleaned_text = remove_arabic_punctuation(text)
print(cleaned_text)
```

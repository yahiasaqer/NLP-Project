```python

import pandas as pd

# Load the original CSV file
file_path = '/content/drive/MyDrive/filtered_shuffled_dataset.csv'
df = pd.read_csv(file_path)

# Split the DataFrame into three parts
chunk_size = 7000
for i in range(3):
    split_df = df[i * chunk_size: (i + 1) * chunk_size]
    split_df.to_csv(f'split_file_{i + 1}.csv', index=False)  # Save each chunk to a new CSV file
```
